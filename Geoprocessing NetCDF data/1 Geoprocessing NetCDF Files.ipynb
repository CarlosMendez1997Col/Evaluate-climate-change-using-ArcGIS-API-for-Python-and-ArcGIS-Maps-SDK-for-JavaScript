## Conversion of units

### Convert units `kg*m-2*s-1` to `mm/month` in precipitation data

The default values of precipitation in NetCDF files are in `kg*m-2*s-1`, for this reason it´s necessary to convert this values to `mm/day` and then in `mm/month`

the constant 86400, convert `kg*m-2*s-1` in `mm/day`. However, the interval slices are 29.5 days. Finally, multiply 86400*29.5 = 2548800, it´s the final constant 

```python
# Model CCCma-CanESM2
with arcpy.EnvManager(scratchWorkspace = r"C:\Users\57320\Dropbox\Mi PC (LAPTOP-5K2HI422)\Documents\ArcGIS for Climate Change\precipitation"):
    out_raster = arcpy.sa.Times(in_raster_or_constant1 = mdRaster_CanESM2_pr, in_raster_or_constant2 = 2548800)
    out_raster.save(r"C:\Users\57320\Dropbox\Mi PC (LAPTOP-5K2HI422)\Documents\ArcGIS for Climate Change\precipitation\CCCma-CanESM2_prcp.crf")
```

```python
# Model MIROC-MIROC5
with arcpy.EnvManager(scratchWorkspace = r"C:\Users\57320\Dropbox\Mi PC (LAPTOP-5K2HI422)\Documents\ArcGIS for Climate Change\precipitation"):
    out_raster = arcpy.sa.Times(in_raster_or_constant1 = mdRaster_MIROC5_pr, in_raster_or_constant2 = 2548800)
    out_raster.save(r"C:\Users\57320\Dropbox\Mi PC (LAPTOP-5K2HI422)\Documents\ArcGIS for Climate Change\precipitation\MIROC-MIROC5_prcp.crf")
```

### Convert `Kelvin` to `Celsius` units in Near Surface Air Temperature Data

The default values of Near Surface Air Temperature in NetCDF files are in `Kelvin`, for this reason it´s necessary to convert this values to `Celsius`. The raster function `Unit Conversion` allow to convert Kelvin in Celsius 


```python
# Model CCCma-CanESM2
from arcpy.ia import *
out_unit_raster = UnitConversion(mdRaster_CanESM2_tas,"Kelvin", "Celsius")
out_unit_raster.save(r"C:\Users\57320\Dropbox\Mi PC (LAPTOP-5K2HI422)\Documents\ArcGIS for Climate Change\temperature\CCCma-CanESM2_tas.crf")
```

```python
# Model MIROC-MIROC5
from arcpy.ia import *
out_unit_raster = UnitConversion(mdRaster_MIROC5_tas,"Kelvin", "Celsius")
out_unit_raster.save(r"C:\Users\57320\Dropbox\Mi PC (LAPTOP-5K2HI422)\Documents\ArcGIS for Climate Change\temperature\MIROC-MIROC5_tas.crf")
```

```python
# Model MOHC-HadGEM2-ES
from arcpy.ia import *
out_unit_raster = UnitConversion(mdRaster_HadGEM2_tas,"Kelvin", "Celsius")
out_unit_raster.save(r"C:\Users\57320\Dropbox\Mi PC (LAPTOP-5K2HI422)\Documents\ArcGIS for Climate Change\temperature\MOHC-HadGEM2-ES_tas.crf")
```
